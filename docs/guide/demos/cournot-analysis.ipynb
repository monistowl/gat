# Cournot Storage Demo Analysis

This notebook plots and inspects the CSV output produced by `bash test_data/demos/storage_cournot.sh`.
It assumes the script was run from the repository root so the data lands in `out/demos/cournot/cournot_results.csv`.

Before running this notebook, execute `bash test_data/demos/storage_cournot.sh` and ensure `out/demos/cournot/cournot_results.csv` exists.


```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
%%matplotlib inline
```

```python
csv_path = Path('out/demos/cournot/cournot_results.csv')
df = pd.read_csv(csv_path)
df.head()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
axes[0].plot(df['N_firms'], df['Price_MWh'], marker='o')
axes[0].set_ylabel('Average Price ($/MWh)')
axes[0].set_title('Market Price vs. Number of Symmetric Storage Firms')
axes[1].plot(df['N_firms'], df['EENS_MWh'], marker='s', color='tab:orange')
axes[1].set_ylabel('Expected EENS (MWh)')
axes[1].set_xlabel('Number of Storage Firms')
axes[1].set_title('Reliability Impact')
plt.tight_layout()
plt.show()
```

*This quick notebook uses pandas/matplotlib to visualize the oligopoly results, matching the reliability narrative from the README.*
